```python
categorical_columns = ['Make', 'Vehicle Class', 'Transmission', 'Fuel Type']
numerical_columns = ['Engine Size(L)', 'Cylinders', 'Fuel Consumption City', 'Fuel Consumption Hwy', 'Fuel Consumption Comb']
bad_categoric = ['not-recorded', 'missing', 'na', 'unspecified', 'unknown', '-1', 'not-available', '-9999', 'zero', 'unestablished']
```

```python
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
import numpy as np
```

```python
for col in df:
    df[col].replace(bad_categoric, np.nan, inplace=True)
    print(df[col].isna().sum())
```

```python
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df = pd.get_dummies(df, categorical_columns)
```
